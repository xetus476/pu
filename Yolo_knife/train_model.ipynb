```python
data_yamal = r"C:\Users\semen\OneDrive\Documents\Ml-itsc\MoneyApp\datasets\knife\data.yaml"

base_model = YOLO("yolov8n.pt")
best_model = YOLO(r"C:\Users\semen\OneDrive\Documents\Ml-itsc\MoneyApp\Yolo_knife\runs\detect\train-6\weights\best.pt")

metrics_base = base_model.val(data=data_yamal, split='val')

metrics_custom = best_model.val(data=data_yamal, split='val')

print("\n" + "="*50)
print(f"{'Метрика':<20} | {'Базовая модель':<15} | {'Дообученная модель':<15}")
print("="*50)
print(f"{'mAP50':<20} | {metrics_base.box.map50:<15.4f} | {metrics_custom.box.map50:<15.4f}")
print(f"{'mAP50-95':<20} | {metrics_base.box.map:<15.4f} | {metrics_custom.box.map:<15.4f}")
print(f"{'Precision (MP)':<20} | {metrics_base.box.mp:<15.4f} | {metrics_custom.box.mp:<15.4f}")
print(f"{'Recall (MR)':<20} | {metrics_base.box.mr:<15.4f} | {metrics_custom.box.mr:<15.4f}")
print(f"{'Inference (ms)':<20} | {metrics_base.speed['inference']:<15.2f} | {metrics_custom.speed['inference']:<15.2f}")
print("="*50)
```
